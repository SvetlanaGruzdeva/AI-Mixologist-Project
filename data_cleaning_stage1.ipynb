```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', 100)
```

```python
df_cocktails = pd.read_csv('./raw_data/all_drinks.csv')
df_cocktails.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_cocktails.head()
```

```python
# Drop obviously excessive columns before detailed review of dataset.

df_cocktails.drop(['dateModified', 'idDrink', 'strDrinkThumb', 'strVideo'], axis=1, inplace=True)
```

```python
df_cocktails.head()
```

```python
df_cocktails.info()
```

```python
df_cocktails['strIBA'].unique()
```

```python
# One more unneccessary column.

df_cocktails.drop('strIBA', axis=1, inplace=True)
```

```python
# Check and fill in missing value in 'Alcoholic' category. It's not possible to remove non-alcoholic rows having Nan in the column.

df_cocktails.loc[df_cocktails['strAlcoholic'].isnull()]
```

```python
df_cocktails['strAlcoholic'].fillna('Alcoholic', inplace=True)
```

```python
# Check how many of non-alcoholic cocktails there are in dataset and drop these rows.

df_cocktails['strAlcoholic'].value_counts()
```

```python
non_alc = df_cocktails.loc[df_cocktails['strAlcoholic'].str.contains('Non')].index
df_cocktails.drop(non_alc, axis=0, inplace=True)
df_cocktails.drop('strAlcoholic', axis=1, inplace=True)
```

```python
# Check and fill in missing instruction.

df_cocktails.loc[df_cocktails['strInstructions'].isna()]
```

```python
df_cocktails['strInstructions'].fillna('Pour 2oz Vodka into a low ball glass, top up with Coke.', inplace=True)
```

```python
# Check 'Ingredient' columns where there are no values in 'Ingredients' but there are some values in 'Measures'.

df_cocktails.loc[df_cocktails['strMeasure13'].notnull()]
```

```python
# Replare blanks with Nan for concistancy.

df_cocktails = df_cocktails.replace(r'^\s+$', np.nan, regex=True)
df_cocktails.info()
```

```python
# Columns related to 13, 14 & 15 ingredient can be dropped.

df_cocktails.drop(['strIngredient13', 'strIngredient14', 'strIngredient15', 'strMeasure13', 'strMeasure14', 'strMeasure15'],
                 axis=1, inplace=True)
df_cocktails.info()
```

```python
# Consolidate all ingredients and all measures in two columns.

df_cons = pd.DataFrame(columns = ['strDrink', 'strCategory', 'strGlass', 'strIngredients','strInstructions', 'strMeasures'])
for i in range(1,13):
    temp = df_cocktails[['strDrink', 'strCategory', 'strGlass', 'strIngredient'+str(i),'strInstructions', 'strMeasure'+str(i)]]
    temp.columns = ['strDrink', 'strCategory', 'strGlass', 'strIngredients','strInstructions', 'strMeasures']
    df_cons = df_cons.append(temp)

# Reset index so it can be used for reference in the future
df_cons.reset_index(inplace=True)
df_cons.drop('index', axis=1, inplace=True)

# Remove lines with no ingredients.
df_cons.drop(df_cons.loc[df_cons['strIngredients'].isnull()].index, axis=0, inplace=True)

# Reset index again so it can be used for reference in the future.
df_cons.reset_index(inplace=True)
df_cons.drop('index', axis=1, inplace=True)

df_cons.shape
```

```python
# Align capitalisation of all ingredients & measures.

df_cons['strIngredients'] = df_cons['strIngredients'].str.title()
df_cons['strMeasures'] = df_cons['strMeasures'].str.lower()
```

```python
# Replace Nan back with '' to be able to use all functionality of pandas.

df_cons.fillna('', inplace=True)
```

```python
# Remove whitespaces in the beginning and in the end of the string.

df_cons['strMeasures'] = df_cons['strMeasures'].apply(lambda x: str(x).strip())
```

```python
df_cons['strMeasures'].unique()
```

```python
# Split of measure column in values and measure names.

df_cons = df_cons.join(df_cons['strMeasures'].str.rsplit(n=1, expand=True).rename(columns={0: 'Value', 1: 'MeasureName'}))
```

```python
df_cons['MeasureName'].unique()
```

```python
# Set up intermediary column for values defined as float.

df_cons[['Value_numeric', 'Value_ml', 'Value_gr', 'Garnish_amount', 'Garnish_type']] = np.nan
df_cons['MeasureName_copy'] = df_cons['MeasureName']
```

```python
df_cons.to_csv('./raw_data/stage1.csv')
```
