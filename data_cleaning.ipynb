```python
# Consolidate all ingredients and all measures in two columns.

df_cons = pd.DataFrame(columns = ['strDrink', 'strCategory', 'strGlass', 'strIngredients','strInstructions', 'strMeasures'])
for i in range(1,13):
    temp = df_cocktails[['strDrink', 'strCategory', 'strGlass', 'strIngredient'+str(i),'strInstructions', 'strMeasure'+str(i)]]
    temp.columns = ['strDrink', 'strCategory', 'strGlass', 'strIngredients','strInstructions', 'strMeasures']
    df_cons = df_cons.append(temp)

# Reset index so it can be used for reference in the future
df_cons.reset_index(inplace=True)
df_cons.drop('index', axis=1, inplace=True)

# Remove lines with no ingredients.
df_cons.drop(df_cons.loc[df_cons['strIngredients'].isnull()].index, axis=0, inplace=True)

# Reset index again so it can be used for reference in the future.
df_cons.reset_index(inplace=True)
df_cons.drop('index', axis=1, inplace=True)

df_cons.shape
```

```python
# Replace Nan back with '' to be able to use all functionality of pandas.

df_cons.fillna('', inplace=True)
```

**Code below is pretty unstructured because original 'Measures' data is very messy and unstructured. So whatever catches an eye first is corrected first.**

\* Note: all individual adjustments are either verified with instructions included in this dataset or with most common recipe available in the Internet.

```python
# Function to print all ingredients with measures and instructions for given cocktail. Comes handy for spot check.

def recipe(name):
    ingredient = []
    measure = []
    for i in range (1, 13):
        ingredient.append(df_cocktails.loc[df_cocktails['strDrink'] == name]['strIngredient'+str(i)].values[0])
        measure.append(df_cocktails.loc[df_cocktails['strDrink'] == name]['strMeasure'+str(i)].values[0])
    df_recipe = pd.DataFrame({'Ingredient': ingredient, 'Measure': measure})
    
    print(df_cocktails.loc[df_cocktails['strDrink'] == name]['strInstructions'].values[0])
    return df_recipe
```

## oz.

```python
# Define oz variable to convert oz to ml.

oz = 30
```

```python
temp = df_cons.loc[(df_cons['MeasureName'] == 'oz')]
temp['Value'].unique()
```

```python
values_replace = {'1/2':0.5, '16':16, '1/3':1/3, '2-3':2.5, '2 1/2':2.5, '2':2, '1':1, '1 1/2':1.5, '3/4':3/4, '3':3,
                   '5':5, '6':6, '1 1/4':1.25, '0.75':0.75, '70ml/2fl':70, '2/3':2/3, '1 2/3':(1+2/3), '1.5':1.5, '4':4,
                   '12':12, '8':8, '1/4':1/4, '0.25':0.25, '0.5':0.5, '28':28, 'Add 10':10, '7':7, '3-4':3.5,'10-12':11}

def repl_from_dic(dic, row):
    if row.Value in dic.keys():
        return dic[row.Value]
    else:
        return row.Value_numeric

# repl_from_dic(values_replace, df_cons.iloc[2])
df_cons['Value_numeric'] = df_cons.loc[(df_cons['MeasureName'] == 'oz')].apply(lambda row: repl_from_dic(values_replace, row),
                                                                               axis=1)
```

```python
df_cons['Value_ml'] = np.where(df_cons['MeasureName'] == 'oz', df_cons['Value_numeric']*oz, df_cons['Value_ml'])
```
