```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
pd.set_option('display.max_rows', None)
```

```python
url = 'https://www.diffordsguide.com/cocktails/search?style=14&sort=rating&limit=80&offset=0'
```

```python
html = requests.get(url).content
html[0:600]
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
cdict = {'class':'box__title'}
table = soup.find_all('h3', cdict)
rating_list = [table[i].contents[0] for i in range(len(table))]
rating_list[:10]
```

```python
# Capitalize each word to align spelling accross the dataset.

rating_list = [rating_list[i].title() for i in range(len(rating_list))]
rating_list[:10]
```

```python
# Remove word 'Cocktail' from the name.

rating_list_clean = [re.sub(r'\ *Cocktail', '', i) for i in rating_list]
rating_list_clean = [re.sub(r'\ *\([^)]*\)\ *', '', i) for i in rating_list_clean]
rating_list_clean[0] = 'Dry Martini'
rating_list_clean[:10]
```

```python
# Review of result. As 100 lines is not so many it worths to be reviewed manually.

pd.DataFrame(zip(rating_list, rating_list_clean))
```

```python
df = pd.DataFrame(rating_list_clean, columns=['Name'])
df['Rating'] = df.index
df.head()
```

```python
df.to_csv('.\\clean_data\\rating.csv')
```
